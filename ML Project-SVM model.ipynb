```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline,Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score
```

```python
#Loading dataset taken from kaggle
cancer_data=pd.read_csv('data.csv')
```

```python
cancer_data.head()
```

```python
# number of rows and Columns in this dataset
cancer_data.shape
```

```python
# checking for missing values
cancer_data.isnull().sum()
```

```python
# checking the distribution of Target Variable
cancer_data['diagnosis'].value_counts()
```

```python
cancer_data.dtypes
cancer_data.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
cancer_data['Outcome']=cancer_data['diagnosis'].map({'M':1,'B':0})
cancer_data.head()
```

1 -->Malignant tumour

2 -->Benign tumour

Separating the features and target









```python
X = cancer_data.drop(columns=['diagnosis','Outcome'], axis=1)
Y = cancer_data['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

```python
 X = np.asarray(X)
 Y = np.asarray(Y)
```

**GridSearchCV**

GridSearchCV is used for determining the best parameters for our model

```python
# loading the SVC model
model = SVC()
```

```python
#Split data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
#Scale data and create pipeline
pipeline = Pipeline([
    ('standardscaler', StandardScaler()),
    ('svc', SVC())
])
```

```python
# hyperparameters

parameters = {
              'svc__kernel':['rbf','linear'],
                'svc__C': [0.1,1,10,100],
              'svc__gamma': [0.001, 0.01, 0.1,1]
}
```

```python
# grid search
classifier = GridSearchCV(pipeline, parameters, cv=5,verbose=3)
```

```python
# fitting the data to our model
classifier.fit(X_train, Y_train)
```

```python
classifier.cv_results_
```

```python
best_para=classifier.best_params_
print(f'The best parameters are:{best_para}')
```

```python

# Training the best model
best_model = classifier.best_estimator_
```

```python
#Predict target
y_predict=best_model.predict(X_test)
```

```python
y_predict
```

```python
# accuracy of the model

accuracy = accuracy_score(Y_test,y_predict)
print(accuracy)
```
